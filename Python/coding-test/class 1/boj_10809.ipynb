# 1. find()와 index()메서드의 차이

## find()
- find(sub, start, end)
    - sub: 찾을 문자열/문자
    - start: 검색 시작 인덱스(생략가능)
    - end: 검색 종료 인덱스(생략가능) - 종료위치는 미포함
- 문자열에서만 사용가능
- 찾는 문자열이 없으면 -1을 반환

## index()
- index(sub, start, end)
    - sub: 찾을 값
    - start: 검색 시작 인덱스(생략가능)
    - end: 검색 종료 인덱스(생략가능) - 종료위치는 미포함
- 문자열, 리스트, 튜플 등 다양한 시퀀스 자료형에서 사용가능
- 찾으려는 값이 없으면 ValueError를 발생시킴

#### 둘다 검색대상을 처음부터 끝까지 순회하면서 탐색

```python
'''
위에 코드는 find() 메서드를 사용하여 각 알파벳의 첫 등장 위치를 찾기에 문자열을 처음부터 끝까지 탐색하는 과정을 26번 반복
해야 하므로 비효율적임 
아래는 한 번의 순회로 문제를 해결하는 더 효율적인 방법
'''
import sys

s = sys.stdin.readline().strip()

# 알파벳 26개를 담을 리스트를 -1로 초기화하여 생성
result = [-1] * 26

# 문자열을 처음부터 끝까지 한 번만 순회합니다.
for i in range(len(s)):
    # ord()를 사용해 문자의 아스키 코드 값을 구한 뒤, 'a'(97)를 빼서 0~25 인덱스로 변환
    idx = ord(s[i]) - 97
    
    # result라는 리스트의 알파벳 위치에 아직 기록이 없을(-1일) 때만 현재 인덱스(i)를 기록하여 첫등장 위치만 기록
    if result[idx] == -1:
        result[idx] = i

print(*result)
```
