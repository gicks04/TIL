```python
# if문에 논리연산자 추가
n = int(input())
if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0):
    t = 1
else:
    t = 0
print(t)
```

```python
# 변수할당에 if문 추가
n = int(input())
t = 1 if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0) else 0
print(t)
```

```python
# 결과값(True/False)을 바로 정수(int)로 변환 -> 왜냐면 true 는 1 false는 0으로 변환됨
n = int(input())
t = int((n % 4 == 0 and n % 100 != 0) or (n % 400 == 0))
print(t)
```
