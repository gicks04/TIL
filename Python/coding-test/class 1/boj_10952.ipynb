## 1. iter() 함수의 두 가지 얼굴
iter() 함수는 사용법에 따라 인자를 1개 또는 2개 받습니다.

- 인자가 1개일 때 (iter(iterable)): 리스트나 문자열 같은 객체를 반복자(Iterator)로 변환합니다.

- 인자가 2개일 때 (iter(callable, sentinel)):

  - 첫 번째 인자(callable): 호출 가능한 객체(함수 이름)를 받습니다.

  - 두 번째 인자(sentinel): 반복을 멈출 **감시값**(중단값)을 받습니다.

## 2. 이터러블(Iterable) vs 이터레이터(Iterator)
- 이터러블 (Iterable): '책' 그 자체입니다. 안에 내용은 많지만, 스스로 "지금 몇 페이지를 읽고 있다"는 정보를 저장하지는 않습니다.

  - 예: 리스트, 문자열, 튜플, 딕셔너리 등

- 이터레이터 (Iterator): 책에 꽂아둔 '책갈피' 혹은 **'손가락'**입니다. "지금 어디까지 읽었는지" 위치를 기억하고, 요청할 때마다 "다음 장(next)"을 보여줍니다.

## 3. 왜 리스트는 이터레이터가 아닐까?
만약 리스트 자체가 이터레이터라면, 리스트 하나를 여러 번의 for 문에서 동시에 돌릴 때 문제가 생깁니다. 책갈피가 하나뿐이라서 여기서 읽으면 저기서도 페이지가 넘어가 버리기 때문이죠.

그래서 파이썬은 다음과 같이 동작합니다.

리스트(Iterable)는 데이터를 보관만 합니다.

for 문이나 iter() 함수를 만날 때, 리스트로부터 새로운 이터레이터(책갈피)를 하나 생성해서 사용합니다.

## 4. Iterator를 통한 for문의 내부 동작
- for 문의 내부 동작 4단계

1. 이터레이터 생성: 가장 먼저 iter([1, 2, 3])을 호출하여 리스트(Iterable)로부터 **이터레이터(Iterator)**라는 책갈피를 만듭니다.

2. 값 요청 (next): 이터레이터에게 next()를 호출하여 "다음 값을 가져와!"라고 요청합니다.

3. 변수 대입: 가져온 값을 x에 담고 for 문 안의 코드를 실행합니다.

4. 반복 및 종료: 더 이상 가져올 값이 없을 때까지 2~3번 과정을 반복하다가, **StopIteration**이라는 특별한 신호(예외)가 발생하면 자동으로 반복을 종료합니다.

```python
# 10952
mlist = []
while True:
  a, b = map(int, input().split())
  mlist.append(a + b)
  if a == 0 and b == 0:
    break
print(*(mlist[:-1]), sep='\n')
# 언패킹 연산자에 슬라이싱한 리스트를 넘겨서 전달하면 원하는 값만 출력가능함
```

```python
# 10952
# '0 0'이 입력될 때까지 계속 input()을 호출합니다.
for line in iter(input, "0 0"):
    # 한 줄씩 가져와서 합계를 바로 출력합니다.
    print(sum(map(int, line.split())))
```
