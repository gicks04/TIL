```python
# 8958
t = int(input())
results = []
for i in range(t):
    string = input()
    score_list = []
    for i in range(len(string)):
        if string[i] == 'O':
            score_list.append(1)
        else:
            score_list.append(0)
    for i in range(len(string)):
        if i > 0 and score_list[i-1] != 0 and score_list[i] != 0:
            score_list[i] = score_list[i-1] + 1
    results.append(sum(score_list))
print(*results, sep='\n')
```

```python
# 점수저장 리스트를 따로 생성하지 않고 현재 연속 점수 를 사용하여 총 점수를 계산

import sys

# 입력을 더 빠르게 받기 위해 sys 사용
for _ in range(int(sys.stdin.readline())):
    string = sys.stdin.readline().strip()
    total_score = 0
    current_consecutive = 0
    
    for char in string:
        if char == 'O':
            current_consecutive += 1      # 성공하면 가산점 증가
            total_score += current_consecutive
        else:
            current_consecutive = 0       # 실패(X)하면 0으로 리셋
            
    print(total_score)
```

# 수학적인 접근 (split 활용)
문자열에서 'X'를 기준으로 문자열을 자르면 'O'가 연속된 묶음들만 남는다는 점을 이용
 1. split('X)'랄 사용하여 X를 기준으로 문자열을 잘라서 잘라진 묶음들로 리스트를 생성
    - 예시: "OOXOOOXO" -> ['OO', 'OOO', 'O']
 2. 묶어진 'O'들의 묶음단위 합은 1~n까지의 합을 구하는 가우스 수열합 공식을 만족한다 
    - 예시: 
        - 'O'가 1개("O")일 때: 1점
        - 'O'가 2개("OO")일 때: 1 + 2 = 3점
        - 'O'가 3개("OOO")일 때: 1 + 2 + 3 = 6점
        - 'O'가 n개일 때: 1+2+⋯+n
 3. 묶어진 'O'의 길이가 n이기에 가우스 수열의 합공식을 적용해서 총점수를 구한다.

```python
import sys

# 입력받은 각 줄에 대해 'X'로 나누어 연속된 'O'들의 길이를 구합니다.
for line in sys.stdin.read().splitlines()[1:]:
    # 'OOO' -> len은 3 -> 점수는 3*(3+1)//2 = 6
    print(sum(len(ox) * (len(ox) + 1) // 2 for ox in line.split('X')))
```
