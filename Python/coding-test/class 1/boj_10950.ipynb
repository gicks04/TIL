```python
# 10950
n = int(input())
mlist=[]
for i in range(n):
  a, b = map(int, input("").split())
  mlist.append(a+b)
print(*mlist, sep ='\n')

# 리스트 컴프리헨션
# 리스트 생성과 입력을 동시에 처리
# mlist = [sum(map(int, input().split())) for _ in range(int(input()))]
# print(*mlist, sep='\n')
# 파이썬의 for 문에서 변수 이름 대신 언더바(_)를 사용하는 것은 "반복은 하되, 그 숫자가 무엇인지는 관심 없다"는 것을 의미
# 파이썬에서 두개의 input함수중에서 반복의 횟수를 먼저 결정하기 위해서 뒤에꺼 input을 먼저 실행
```
