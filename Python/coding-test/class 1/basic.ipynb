## 🐍 파이썬 리스트 컴프리헨션 (List Comprehension)
리스트 컴프리헨션은 기존의 반복 가능한(Iterable) 객체를 활용하여 새로운 리스트를 간결하게 생성하는 파이썬의 핵심 문법
### 1. 기본 구조 (추출 및 변형)
가장 기본적인 형태로, 요소를 하나씩 꺼내어 리스트에 담습니다.
- 구문: [ 표현식 for 항목 in Iterable ]
- 예시: 숫자를 2배로 만들기
  - numbers = [10, 20, 30, 40, 50]
  - doubled = [x * 2 for x in numbers]
  - 결과: [20, 40, 60, 80, 100]
### 2. 필터링형 (if 조건문)
if 문을 뒤에 붙여 조건이 **참**(True)인 요소만 골라냅니다.
- 구문: [ 표현식 for 항목 in Iterable if 조건식 ]
- 예시: 'a'가 포함된 과일만 추출하기
  - fruits = ["apple", "banana", "cherry", "kiwi"]
  - has_a = [f for f in fruits if "a" in f]
  - 결과: ['apple', 'banana']
### 3. 조건부 변형형 (if-else 조건문)
값에 따라 다른 결과를 담고 싶을 때 사용합니다.이때는 if-else가 for문 앞에 위치합니다.
- 구문: [ 표현식1 if 조건식 else 표현식2 for 항목 in Iterable ]
- 예시: 짝수는 'Even', 홀수는 'Odd'로 표시하기
  - labels = ["Even" if x % 2 == 0 else "Odd" for x in range(5)]
  - 결과: ['Even', 'Odd', 'Even', 'Odd', 'Even']
### 4. 중첩 반복문 (다중 for문)
여러 개의 반복문을 한 줄에 중첩하여 사용할 수 있습니다.
- 구문: [ 표현식 for 항목1 in Iterable1 for 항목2 in Iterable2 ]
- 예시: 두 리스트의 모든 조합 곱하기
  - list1 = [2, 3]
  - list2 = [1, 2, 3]
  - multiplied = [i * j for i in list1 for j in list2]
  - 결과: [2, 4, 6, 3, 6, 9]
### 5. 중첩 리스트 생성 (2차원 배열)
리스트 안에 또 다른 리스트 컴프리헨션을 넣어 행렬 구조를 만듭니다.
- 구문: [ [표현식 for 항목1 in Iterable1] for 항목2 in Iterable2 ]
- 예시: 3x2 모양의 영행렬 만들기
- matrix = [[0 for _ in range(2)] for _ in range(3)]
- 결과: [[0, 0], [0, 0], [0, 0]]

## 1. iter() 함수의 두 가지 얼굴
iter() 함수는 사용법에 따라 인자를 1개 또는 2개 받습니다.

- 인자가 1개일 때 (iter(iterable)): 리스트나 문자열 같은 객체를 반복자(Iterator)로 변환합니다.

- 인자가 2개일 때 (iter(callable, sentinel)):

  - 첫 번째 인자(callable): 호출 가능한 객체(함수 이름)를 받습니다.

  - 두 번째 인자(sentinel): 반복을 멈출 **감시값**(중단값)을 받습니다.

## 2. 이터러블(Iterable) vs 이터레이터(Iterator)
- 이터러블 (Iterable): '책' 그 자체입니다. 안에 내용은 많지만, 스스로 "지금 몇 페이지를 읽고 있다"는 정보를 저장하지는 않습니다.

  - 예: 리스트, 문자열, 튜플, 딕셔너리 등

- 이터레이터 (Iterator): 책에 꽂아둔 '책갈피' 혹은 **'손가락'**입니다. "지금 어디까지 읽었는지" 위치를 기억하고, 요청할 때마다 "다음 장(next)"을 보여줍니다.

## 3. 왜 리스트는 이터레이터가 아닐까?
만약 리스트 자체가 이터레이터라면, 리스트 하나를 여러 번의 for 문에서 동시에 돌릴 때 문제가 생깁니다. 책갈피가 하나뿐이라서 여기서 읽으면 저기서도 페이지가 넘어가 버리기 때문이죠.

그래서 파이썬은 다음과 같이 동작합니다.

리스트(Iterable)는 데이터를 보관만 합니다.

for 문이나 iter() 함수를 만날 때, 리스트로부터 새로운 이터레이터(책갈피)를 하나 생성해서 사용합니다.

## 4. Iterator를 통한 for문의 내부 동작
- for 문의 내부 동작 4단계

1. 이터레이터 생성: 가장 먼저 iter([1, 2, 3])을 호출하여 리스트(Iterable)로부터 **이터레이터(Iterator)**라는 책갈피를 만듭니다.

2. 값 요청 (next): 이터레이터에게 next()를 호출하여 "다음 값을 가져와!"라고 요청합니다.

3. 변수 대입: 가져온 값을 x에 담고 for 문 안의 코드를 실행합니다.

4. 반복 및 종료: 더 이상 가져올 값이 없을 때까지 2~3번 과정을 반복하다가, **StopIteration**이라는 특별한 신호(예외)가 발생하면 자동으로 반복을 종료합니다.

      - %S: 앞의 빈자리를 0으로 채우는 2자리 초 숫자

      - %B: 영어로 된 월 문자열
    - **parsing 방법**
      - 데이터: "2026 01 13" -> 포맷: "%Y %m %d"

## sort
- sort는 list객체에 있는 메서드
- 원본리스트를 직접수정

    - 정렬 기준을 정하는 함수를 인자로 받습니다.
    - 예시: key=len을 넣으면 문자열의 길이를 기준으로 정렬
- 예시: mylist.sort() or mylist.sort(reverse=True)

## sorted
- sorted는 내장함수
- 원본리스트를 수정하지 않고 정렬된 새 리스트를 생성
- 정렬된 새 리스트를 반환함
- if문의 조건에 사용가능
- 매개변수
  - iterable (필수):
    - 첫 번째 인자로 리스트, 튜플, 딕셔너리, 문자열 등 정렬할 대상
  - reverse (bool):
    - .sort()와 동일하게 True일 경우 내림차순 정렬
  - key (callable):
    - 정렬의 기준이 되는 함수를 지정
- 예시 new_list = sorted(mylist) or sorted(mylist, reverse=True)
