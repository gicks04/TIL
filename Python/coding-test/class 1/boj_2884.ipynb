# datetime 패키지
## 1. datetime 객체
날짜와 시간정보를 모두 가진 객체
- 멤버
  - year: 연도
  - month: 월
  - day: 일
  - hour: 시
  - minute: 분
  - second: 초
  - microsecond: 마이크로초(micro seconds, 백만분의 일초)

 `x.year, x.month, x.day, x.hour, x.minute, x.second, x.microsecond`

- 메서드
  - datetime.date(): datetime객체를 날짜정보만 가진 date객체로 반환
  - datetime.time(): datetime객체를 시간정보만 가진 time객체로 반환
  - datetime.combine(date, time): 별도의 날짜 객체와 시간 객체를 하나로 합칩니다.
  - **datetime.now**(): 현재날짜와 시간을 datetime객체로 반환
  - datetime.weekday(): 요일 반환 (0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일)
  - **datetime.strftime**(): datetime객체를 원하는 형식의 문자열로 formatting
  - **datetime.strptime**(): 문자열을 원하는대로 parsing하여 datetime 객체로 변환

    - **날짜 formatting, parsing 문자열**
      - %Y: 앞의 빈자리를 0으로 채우는 4자리 연도 숫자
      - %m: 앞의 빈자리를 0으로 채우는 2자리 월 숫자
      - %d: 앞의 빈자리를 0으로 채우는 2자리 일 숫자
      - %H: 앞의 빈자리를 0으로 채우는 24시간 형식 2자리 시간 숫자
      - %M: 앞의 빈자리를 0으로 채우는 2자리 분 숫자
      - %S: 앞의 빈자리를 0으로 채우는 2자리 초 숫자
      - %A: 영어로 된 요일 문자열
      - %B: 영어로 된 월 문자열
    - **parsing 방법**
      - 데이터: "2026 01 13" -> 포맷: "%Y %m %d"
      - 데이터: "2026-01-13 16:20" -> 포맷: "%Y-%m-%d %H:%M"
      => 보이는 바와 같이 formatting문자열들 사이에 있는 문자가 구분자가 되어서 parsing함
      => 출력할때도 마찬가지로 formatting문자열들은 숫자가 들어가서 문자열이 완성됨(f스트링느낌)



```python
# 2884
from datetime import datetime, timedelta
a, b = map(int, input().split())
dt = datetime(2026, 1, 1, a, b)
result = dt - timedelta(minutes=45)
print(result.hour, result.minute)
```
