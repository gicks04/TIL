# sort와 sorted
## sort
- sort는 list객체에 있는 메서드
- 원본리스트를 직접수정
- 반환값이 None -> 원본리스트를 직접 정렬해두기 때문에
- 반환값이 없기때문에 if문의 조건에 직접적으로 사용불가
- 매개변수
  - reverse (bool):
    - 기본값은 False(오름차순)
    - True로 설정하면 내림차순정렬

  - key (callable):
    - 정렬 기준을 정하는 함수를 인자로 받습니다.
    - 예시: key=len을 넣으면 문자열의 길이를 기준으로 정렬
- 예시: mylist.sort() or mylist.sort(reverse=True)

## sorted
- sorted는 내장함수
- 원본리스트를 수정하지 않고 정렬된 새 리스트를 생성
- 정렬된 새 리스트를 반환함
- if문의 조건에 사용가능
- 매개변수
  - iterable (필수):
    - 첫 번째 인자로 리스트, 튜플, 딕셔너리, 문자열 등 정렬할 대상
  - reverse (bool):
    - .sort()와 동일하게 True일 경우 내림차순 정렬
  - key (callable):
    - 정렬의 기준이 되는 함수를 지정
- 예시 new_list = sorted(mylist) or sorted(mylist, reverse=True)

```python
# 2920
mclist = list(map(int, input().split()))
if mclist == sorted(mclist):
  print('ascending')
elif mclist == sorted(mclist, reverse=True):
  print('descending')
else:
  print('mixed')
```
