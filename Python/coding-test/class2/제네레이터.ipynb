# 1. 제네레이터란?

### 일반적인 리스트완 달리 모든 데이터를 메모리에 미리 저장해두지 않음 -> 필요할때만 생성
#### 1. 제네레이터생성(yield 키워드 사용)
- generator: itrator를 생성해주는 함수, 함수 안에 yield 키워드를 사용함
- 특징
    - iterable한 순서가 지정됨
    - 다음값을 무조건 갖고오지 않고 필요하면 가져옴
    - 함수 내부 로컬별수를 통해 내부상태가 유지(순서가 유지됨)
    - 더이상 생성할 수 없다면 StopIteration 에러를 발생
    - 한번 끝까지 읽으면 데이터가 사라짐 -> 재사용 어려움
```python
>>> def test_generator():
...     yield 1
...     yield 2
...     yield 3

# 결과 출력
>>> gen = test_generator()
>>> type(gen)
<class 'generator'>
>>> next(gen)
1
>>> next(gen)
2
>>> next(gen)
3
>>> next(gen)
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
StopIteration
```
    - generator를 두개 생성하면, 서로 다른 객체이기 때문에 각기 따로 동작
```python
>>> h = test_generator()
>>> i = test_generator()
>>> h == i
False
>>> h is i
False
>>> next(h)
1
>>> next(i)
1
>>> next(h)
2
>>> next(i)
2
>>> next(i)
3
>>> next(h)
3
```

#### 2. 제네레이터 생성(제네레이터 표현식)
리스트 컴프리헨션과 비슷하지만 소괄호() 사용
 ```python3
 # 리스트 컴프리헨션과 달리 계산규칙만 가지고 피요할때마다 생성
 ㅡmy_gen = (x for x in range(1000000))
 ```

````python
# 30802 - 제네레티어를 사용하여 코드를 간결
n = int(input())
tlist = map(int, input().split())
t, p = map(int, input().split())

# 수식을 제너레이터 표현식으로 sum()에 전달
print(sum((size - 1) // t + 1 for size in tlist))

print(n // p, n % p)
````
